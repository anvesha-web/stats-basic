```python
#Practical questions
#3 
import statistics as stats

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean_val = stats.mean(numbers)

# Median
median_val = stats.median(numbers)

# Mode
mode_val = stats.mode(numbers)

print("Numbers:", numbers)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
#4
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Covariance matrix
cov_matrix = np.cov(x, y, bias=False)   # bias=False â†’ sample covariance
cov_xy = cov_matrix[0, 1]

# Correlation coefficient
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

# Output
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance (X, Y):", cov_xy)
print("Correlation Coefficient (X, Y):", corr_xy)
```

```python
#5
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# Calculate IQR to identify outliers
Q1 = np.percentile(data, 25)   # First quartile (25th percentile)
Q3 = np.percentile(data, 75)   # Third quartile (75th percentile)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
#6
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Covariance matrix
cov_matrix = np.cov(x, y, bias=False)
cov_xy = cov_matrix[0, 1]

# Correlation coefficient
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

# Output
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance (X, Y):", cov_xy)
print("Correlation Coefficient (X, Y):", corr_xy)
```

```python
#7
import matplotlib.pyplot as plt
import statistics as stats

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean_val = stats.mean(survey_scores)
median_val = stats.median(survey_scores)
mode_val = stats.mode(survey_scores)
std_dev = stats.pstdev(survey_scores)  # population standard deviation
min_val, max_val = min(survey_scores), max(survey_scores)

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
print("Standard Deviation:", std_dev)
print("Minimum:", min_val)
print("Maximum:", max_val)

# Histogram
plt.hist(survey_scores, bins=6, color="skyblue", edgecolor="black")
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Scores (1â€“10)")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
